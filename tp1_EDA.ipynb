### Variables que si serán tomadas en cuenta

•**Year:** año de la temporada que se jugó. **No es de entrada**, pero la vamos a utilizar a la hora de entrenar el modelo<br>
•**Player:** nombre del jugador. **No es de Entrada**, pero la vamos a utilizar a la hora de entrenar el modelo<br>
•**Age:** edad del jugador.  **Entrada**. La vamos a utilizar porque creemos que es importante la edad cuando hablamos del rendimiento del jugador en la temporada siguiente.<br>
•**MP:** “Minutes Played” son los minutos en cancha del jugador en la temporada. Si, **entrada**. Es necesaria ya que si un jugador juega más minutos tiene más oportunidades de meter mas puntos.<br>
•**PER:** “Player Efficiency Rating” es una métrica que suma todos los logros positivos que tuvo el jugador y le resta lo negativo. El resultado es una calificación por minuto del desempeño del jugador. Si, **entrada**. Es un indicador muy bueno de lo que hizo el jugador en la temporada.<br>
•**TS%:** “True Shooting Percentage” es una medida de la eficiencia de tiro que tiene en cuenta los tiros de campo, los tiros de campo de 3 puntos y los tiros libres. Si, **entrada**. Muy importante, ya que posee gran relevancia a la hora de medir la cantidad de puntos que puede lograr un jugador<br>
•**ORB%:** “Offensive Rebound Percentage” se refiere al porcentaje de rebotes ofensivos, y se trata de una estimación del porcentaje de rebotes ofensivos que un jugador agarró mientras estaba en la cancha. Básicamente es una medida de cuantos rebotes en ataque agarró el jugador. SI, **entrada**. Pensamos que mientras mas rebotes en ataque haya agarrado el jugador, más oportunidades tiene de realizar puntos.<br>
•**STL%:** “Steal Percentage” o el porcentaje de robos. Es una estimación de la cantidad de veces que el jugador le robó la pelota a un jugador contrario. SI, **entrada**. Creemos que cuando el jugador le roba la pelota a un jugador del equipo contrario, tiene más posibilidades de terminar la jugada haciendo algún punto.<br>
•**TOV%:** “Turnover Percentage” o el porcentaje de pérdidas. Pérdidas cometidas por cada 100 posesiones. SI, **entrada**. Creemos que las veces que el jugador perdió la pelota influyen negativamente en la cantidad de puntos que puede realizar.<br>

•**2P%:** Two Points - Dos Puntos: es el porcentaje de tiros de 2 puntos anotados. Si, **entrada**. Puede influir a la hora de calcular los puntos, ya que, si tiene un gran porcentaje de tiros de 2 puntos anotados, probablemente la temporada siguiente haga más puntos<br>
•**eFG%:** porcentaje efectivo de tiros de campo anotados, es decir, es la cantidad de tiros de campo anotados, ponderados según su valor respecto a los tiros de campo intentados. SI, **entrada**, ya que consideramos que es una variable importante y a tener en cuenta<br>
•**FT%:** es el porcentaje de tiros libres anotados. Si, **entrada**, puede influir a la hora de calcular los puntos, ya que, si tiene un gran porcentaje de tiros libres anotados, probablemente la temporada siguiente haga más puntos<br>
•**PTS:** “Points” o puntos totales realizados por el jugador en la temporada. Si, **entrada**. Los puntos son muy importantes ya que con esta columna podemos definir nuestra variable target.<br>
•**USG%:** Usage Percentage - Porcentaje de Uso. Intenta medir la importancia de un jugador en el ataque de su equipo. SI, **entrada**.  Creemos que es una variable muy importante, ya que, si el equipo depende mucho del jugador, va a tener mas posibilidades de hacer puntos.<br>
•**WS:** Offensive Win Shares - Acciones de victoria ofensivas. Es una métrica que estima la cantidad de victorias que un jugador produce para su equipo debido a su habilidad ofensiva. SI, **entrada**. Creemos que si un jugador tiene un bajo valor OWS quiere decir que no es tan relevante en el ataque, y por ende va a realizar menos puntos. Es un dato que creemos relevante para tener en cuenta.<br>
•**OBPM:** Offensive Box Plus-Minus - Caja de Mas-Menos Ofensiva: Mide la aportación de un jugador al rendimiento total del equipo a través de sus estadísticas por cada 100 posesiones ofensivas. SI, **entrada**. Porque tiene gran relevancia para calcular el target.<br>
•**VORP:** Value Over Replacement Player: esta estadística mide el rendimiento de un jugador contra el rendimiento que daría un “jugador medio” de la liga. SI, **entrada**. Creemos que, si un jugador resalta mas que el promedio, es mas probable que haga más puntos<br>

•**FG%:** Field Goals - Tiros de campo: es el porcentaje de tiros de campo anotados. Si, **entrada**. Puede influir a la hora de calcular los puntos, ya que, si tiene un gran porcentaje de tiros de campo anotados, probablemente la temporada siguiente haga más puntos<br>
•**3P%:** Three Points - Tres Puntos: es el porcentaje de tiros de 3 puntos anotados. Si, **entrada**. Puede influir a la hora de calcular los puntos, ya que, si tiene un gran porcentaje de tiros de 3 puntos anotados, probablemente la temporada siguiente haga más puntos<br>

```python
# Por lo tanto para el jugador Ron Brewer (por ejemplo), nos quedaríamos con sus "totales" por temporada.
jugador = df[df['Player'] == 'Andres Nocioni']
display(jugador) 
```

```python
# Esta celda renombra la primer columna del dataFrame (denominada 'Unnamed: 0') y le setea el nombre de "id"
df = df.rename(columns = {'Unnamed: 0' : 'id'})
```

```python
# Con esta celda, creamos la que será nuestra columna TARGET y la llamamos "morePoints" ("Mas Puntos")
# Dicha columna podrá tener solo dos valores "Yes"/"No" dependiendo si el jugador hizo mas puntos que la temporada anterior
# Cabe recalcar que si es la 1er temporada del jugador o si una temporada no jugo por algun motivo, se colocará en "Yes" ya que consideramos que al no haber jugado efectivamente realizó mas puntos.
df.insert(18, 'morePoints', 'Yes', allow_duplicates = True)
```

```python
# Esta es una funcion que realizamos para setear la columna "morePoints"
# Buscamos el año anterior y comparamos los puntos que realizó en ese año y el año actual.
# Si efectivamente hizo mas puntos no hacemos nada, pero si no seteamos la columna en "No"
def funcionSetMorePoints(id):
    auxiliar = df.loc[id]
    temporadaAnterior = df.loc[(df.Player == auxiliar.Player) & (df.Year == auxiliar.Year  - 1)]      
    if ((not temporadaAnterior.empty) and (auxiliar.PTS <= temporadaAnterior.PTS.item())):
        df.loc[df.id == id,'morePoints'] = 'No'
    return id
```

```python
# Llamamos a la funcion defenida anteriormente
df.id = df.id.apply(funcionSetMorePoints)
```

```python
# Con esta linea de codigo podemos buscar cualquier jugador y revisar si la funcion anterior seteó los datos correctamente.
jugador = df[df['Player'] == 'Manu Ginobili']
jugador[['Year','Player','Age','PTS','morePoints']]
```
